```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
df=pd.read_csv('EPL_Soccer_MLR_LR.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Exploratory Analysis to determin predictor variable
df.corr()
```

```python
#Plot cost vs. score

plt.scatter(df['Cost'], df['Score']);
```

```python
#Strong linear association between cost and score!
```

```python
#Assisgn X, Y, use train_test_split

x=df['Cost']
y=df['Score']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75,
                                                   test_size= 0.25, random_state=100);
```

```python
#statsmodel approach to regression
#fit the model

lr = sm.OLS(y_train, x_train).fit()
```

```python
#Printing the parameters
lr.params
lr.summary()
```

```python
#Force intercept term
x_train_with_intercept = sm.add_constant(x_train)
lr = sm.OLS(y_train, x_train_with_intercept).fit()
lr.summary()
```

```python
#Extract the B0, B1
b0 = lr.params[0]
b1 = lr.params[1]
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, b0+b1*x_train, 'r')
plt.show()
```

```python
#Plot the fitted line on the test data
x_test_with_intercept = sm.add_constant(x_test)
y_test_fitted = lr.predict(x_test_with_intercept)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_test_fitted, 'r')
plt.show()
```

```python
#CHECKLIST
#Non-Linearity
#Non-Constant Variance
#Deviations from Normality
#Errors Not IID
#Outliers
#Missing Predictions
```

```python
##Build Predictions on training data - Non-linearity
predictions_y = lr.predict(x_train_with_intercept)
```

```python
#Find residuals
r_i = (y_train - predictions_y)
```

```python
plt.title('Residuals vs. Cost')
plt.xlabel('Cost',fontsize=15)
plt.scatter(x_train, r_i)
plt.show()
```

```python
#Absolute residuals against predictor
abs_r_i = np.abs(y_train - predictions_y)
plt.title('Absolute Residuals vs. Cost')
plt.xlabel('Cost', fontsize=15)
plt.scatter(x_train, abs_r_i)
plt.show()
```
